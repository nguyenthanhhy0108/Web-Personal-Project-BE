# Ensure you insert all brand into your MySQL database `brand` column in `df` and please insert `porsche` manually

```python
url = 'http://localhost:8080/vehicle-inventory/vehicles'

for index in range(len(df)):
    if index == 0:
        continue
    row = df.iloc[index]

    vehicleBrandName = row['5']
    vehicleName = row['1']
    vehiclePrice = row['3']
    vehicleImageUrl = row['4']
    numberOfRemaining = 10
    vehicleDescription = row['2']

    data = {
        "vehicleBrandName" : vehicleBrandName,
        "vehicleName" : vehicleName,
        "vehiclePrice" : int(vehiclePrice),
        "vehicleImageUrl" : vehicleImageUrl,
        "numberOfRemaining" : numberOfRemaining,
        "vehicleDescription" : vehicleDescription
    }

    headers = {
        'Content-Type': 'application/json',
    }

    response = requests.post(url, headers=headers, data=json.dumps(data))

    if response.status_code == 200:
        print("Success!")
        print("Response:", response.json())
    else:
        print(f"Fail, status code: {response.status_code}")
        print("Response:", response.text)

    # print(data)
```
